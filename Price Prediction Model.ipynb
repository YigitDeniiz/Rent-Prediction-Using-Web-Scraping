## Gereksiz ve Yanlış Yazılmış Olanları Çıkartalım: 

```python
df = df.drop("tarih",axis=1)  #tarihi kullanmayacağız çıkartıyorum.
for i in range(len(df)):
    df.fiyat[i] = df.fiyat[i]*1000  #fiyatı 5 haneli haline geri çeviriyorum.
```

```python
for i in range(len(df)):
    if df.boyut[i] == "135 + 105": #sıkıntılı olan tek boyut bilgisini çıkartıyoruz.
        df = df.drop(i)
df = df.reset_index()
```

```python
for i in range(len(df)):  
    if len(df.kat[i])>9:    # kat bilgisi yerine başka bir şey yazıldıysa çıkartıyoruz.
        df = df.drop(i)
df = df.reset_index()
```

```python
for i in range(len(df)):
    if len(df.adres[i])<15: # adres bilgisinde kısa olup adres geçmeyenleri çıkartıyoruz.
        df = df.drop(i)
```

```python
df = df.drop("level_0",axis=1) #indexleri 0 larken oluşmuş yeni kolonları çıkartıyorum.
df = df.drop("index",axis=1)
```

Şimdi verimizde istemediğimiz veya yanlış yazılmış veri kalmadı. Şimdi veriyi tanıyıp düzenleme kısmı var.

## Data Preprocess:

```python
df.head() 
```

## Boyut kısmına data mapping yapalım.

```python
df.boyut.value_counts() 
```

```python
daire = {"Stüdyo":0,"1 + 1":1,"2 + 1":2,"2 + 2":3,"3 + 1":4,"4 + 1":5,"4 + 2":6,"6 + 1":7}
df["boyut"] = df["boyut"].map(daire)
df.head()
```

M2 ve yaş sütunları için; "m2" "Yaşında" kelimelerini çıkartalım ki sadece int değerler kalsın.

```python
df['m2'] = df['m2'].map(lambda x: x.rstrip('m²'))
df['yaş'] = df['yaş'].map(lambda x: x.rstrip('Yaşında'))
df = df.replace("Sıfır Bi",0) #Sıfır Bina olanları 0'a eşitledim.
df["m2"] = df["m2"].astype(int)
df["yaş"] = df["yaş"].astype(int)
df.head()
```

### Kat kısmını düzenleyelim:

```python
df.kat.value_counts()
```

```python
df = df.replace("Bodrum",-1)
df = df.replace("Zemin",0)
df = df.replace("Ara Kat",5)
df = df.replace("Kot 1",-2)
df = df.replace("Kot 2",-2)
df = df.replace("Kot 3",-2)
df = df.replace("Çatı Katı",21)
df["kat"] = df["kat"].astype(str)
df['kat'] = df['kat'].map(lambda x: x.rstrip('. Kat'))
df["kat"] = df["kat"].astype(int)
df["kat"] = df["kat"] + 2
df.kat.value_counts()
```

### Adres Bilgisi Sınıflandırma:

```python
df.adres.value_counts()
```

```python
dummies = pd.get_dummies(df[["adres"]],drop_first=True)
df = df.drop(["adres"],axis=1)
df = pd.concat([df,dummies],axis="columns")
df.head()
```

## Modelleri Deneyip En İyi Sonuç Alacağımız Modeli Seçelim:

```python
from lazypredict.Supervised import LazyClassifier
from lazypredict.Supervised import LazyRegressor
from sklearn.model_selection import train_test_split
```

```python
y = df.fiyat
X = df.drop(["fiyat"],axis=1)

X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.2,random_state=12)

reg = LazyRegressor(verbose=0,ignore_warnings=True, custom_metric=None )
models,predictions = reg.fit(X_train, X_test, y_train, y_test)
models
```

Regresyon modelleri arasında en iyi sonucu GradientBoostingRegressor modelinde %82 başarı ile sağlıyoruz.

```python
from sklearn.ensemble import GradientBoostingRegressor
reg2 = GradientBoostingRegressor(random_state=0)
reg2.fit(X_train, y_train)
reg2.score(X_test, y_test)
```

Fiyatı tahmin edilecek dairenin m2,yaş,kat ve mahallesinin girildiği bir dataframe ile reg2.predict(tahmin) kullanarak %82 başarı oranı ile tahmin edilebilir.
